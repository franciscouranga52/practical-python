````python
## import libraries
import glob
import pandas as pd
````

````python
## grab only the txt files
all_txts = sorted(glob.glob ("project-docs/*.txt"))
all_txts
````

````python
## create a list of dictionaries that holds the year, cognition_related , medical_condition, care_hours
dictionary_list = []
for txt_file in all_txts:
    with open(txt_file, "r") as my_text:
        line = my_text.readlines()
        year = line[0].replace("Year: ", "").replace("\n","")
        if "not suffer" in line[2]:
            cognition_related = "False"
        else: 
            cognition_related = "True"
        if "Dementia" in line[2]:
            medical_conditions = "Dementia"
        else:
            if "Alzheimer's disease" in line[2]:
                medical_conditions = "Alzheimer's disease"
            else:
                medical_conditions = "Not Specified"
        if "not suffer" in line[2]:
            decision = "False"
        else: 
            decision = "True"
        if "12-hour" in line[4]:
            care_hours = "Half-day"
        else:
            if "24-hour" in line[4]:
                care_hours = "Full-day"
            else:
                care_hours = "Not Specified"
        source = txt_file
        decisions_dict = {"Year" : year, 
                          "Cognition related" : cognition_related, 
                            "Medical conditions" : medical_conditions, 
                            "Care hours" : care_hours,
                            "Source" : source}
        dictionary_list.append(decisions_dict)
````

````python
df = pd.DataFrame (dictionary_list)
df
````

````python
## use pandas to write to csv file
df.to_csv("diagnosis.csv", encoding = "UTF-8", index = False )
````
