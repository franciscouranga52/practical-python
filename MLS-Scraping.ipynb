```python
#import libraries]
import pandas as pd
import requests # es lo que trae todo el html del servidor
from bs4 import BeautifulSoup
import time 
from random import randrange
import icecream
```

```python
#import json of matches
matches_url = "https://sportapi.mlssoccer.com/api/matches?culture=en-us&dateFrom=2011-03-01&dateTo=2023-11-01"
```

```python
response = requests.get(matches_url)
response.status_code
```

```python
data_matches = response.json()
```

```python
id_list = [data_matches[i]["optaId"] for i in range(0,len(data_matches))]
match_days = [data_matches[i]["matchDate"] for i in range(0,len(data_matches))]
id_list
```

```python
len(id_list)
```

```python
len(match_days)
```

```python
# import json of every match
start_url = "https://stats-api.mlssoccer.com/v1/matches?&match_game_id=" 
end_url = "&include=away_club_match&include=home_club_match&include=venue&include=home_club&include=away_club"
url_list = [f"{start_url}{id}{end_url}" for id in id_list]
url_list
```

```python
len(url_list)
```

```python
dict_list = []
broken_url = []
counter = 1
for url in url_list:
    try:
        print(f"Scraping {counter} of {len(url_list)}")
        response = requests.get(url)
        data = response.json()
        #print(data)
        date = match_days[counter-1]
        counter +=1
        weather = data[0] ["weather"] 
        stadium = data[0] ["venue"]["name"]
        city= data[0] ["venue"]["city"]
        capacity = data[0] ["venue"]["capacity"]
        attendance = data[0] ["attendance"]
        away_club = data[0] ["away_club"] ["name"]
        home_club = data[0] ["home_club"] ["name"]
        away_club_score = data[0] ["away_club_match"] ["score"]
        home_club_score = data[0] ["home_club_match"] ["score"]
        match_type = data[0] ["type"]
        source = url
        
        if data[0] ["home_club_match"]["is_winner"] == True:
            result = "home club wins"
        else:
            if data[0] ["away_club_match"]["is_winner"] == True:
                result = "away club wins"
            else:
                result = "tie"
        dictionary = {
            "Date": date,
            "Weather" : weather,
            "Stadium" : stadium,
            "City" : city,
            "Capacity" : capacity,
            "Attendance" : attendance,
            "Away club" : away_club,
            "Home club" : home_club,
            "Away club score" : away_club_score,
            "Home club score" : home_club_score,
            "Match result" : result,
            "Match type": match_type,
            "Source": source
        }
        dict_list.append(dictionary)
    except:
        print(f"Match {date} between {home_club} and {away_club} has no ID")
        broken_url.append(url)
    finally:
        snooze = randrange (2,7)
        print(f"Snoozing for {snooze} seconds")
        time.sleep (snooze)
print ("Done scraping all units")
```
